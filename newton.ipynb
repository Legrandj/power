```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Newton-Raphson iteration for z**3-1=0
def phi(z):
    z=complex(z)
    return z+(1/(z*z)-z)/3
#cubic roots of 1
zz=[1,complex(-.5,.5*np.sqrt(3)),complex(-.5,-.5*np.sqrt(3))]
#corresponding colors
cc=[[1,0,0],[0,1,0],[0,0,1]]
```

```python
xmin=-3.
xmax=3.
ymin=-3.
ymax=3.
eps=.001
maxit=100
imgx=512
imgy=512
img=np.zeros((imgy,imgx,3))
a=(xmax-xmin)/imgx
b=(ymax-ymin)/imgy
for i in range(imgy):
    for j in range(imgx):
        x=xmin+a*j
        y=ymax-b*i
        z0=complex(x,y)
        for k in range(maxit):
            if(abs(z0)<eps):
                break
            z=phi(z0)
            con=False
            for l in range(len(zz)):
                if(abs(z-zz[l])<eps):
                    con=True
                    break
            if(con):
                img[i,j]=cc[l]
                break
            z0=z
plt.figure(figsize=(10,10))
plt.imshow(img,extent=[xmin,xmax,ymin,ymax])
plt.show()
```
