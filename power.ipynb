```python
import numpy as np
import matplotlib.pyplot as plt
import random as rn
```

```python
def expser(n,x,e=0):
#truncated Maclaurin series (i.e., Taylor polynomial) for exp
# n - degree of the Taylor polynomial
# x - the argument
# e - relative error in computation of the series terms, modeled by uniform noise
    b=1.
    y=b
    for i in range(1,n+1):
        b*=float(x)/i
        r=2*e*rn.random()-e
        y+=b*(1.+r)
    return y
```

```python
def expserlist(n,xlist,e=0):
#truncated Maclaurin series (i.e., Taylor polynomial) for exp
# n - degree of the Taylor polynomial
# x - the argument *list*
# e - relative error in computation of the series terms, modeled by uniform noise
    ylist = []
    for x in xlist:
        ylist.append(expser(n,x,e))
    return ylist
```

```python
def err(n,x,elist):
    yy = []
    for e in elist: 
        yy.append(expser(n,x,e)-np.exp(x))
    return yy
```

```python
xlist = np.linspace(-3, 2, 50)
elist = np.logspace(-3, -.5, 150)
nlist = [3,4,16]

f,(ax1,ax2) = plt.subplots(2)
ax1.plot(xlist,np.exp(xlist))
for n in nlist: 
    ax1.plot(xlist,expserlist(n,xlist,.1))
for n in nlist:
    ax2.plot(elist,err(n,2.,elist))
ax2.set_xscale('log')

plt.show()
```
