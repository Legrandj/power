```python
import numpy as np
import matplotlib.pyplot as plt
from random import *
```

```python
def expser(n,x,e=0):
#truncated Maclaurin series (i.e., Taylor polynomial) for exp
# n - degree of the Taylor polynomial
# x - the argument
# e - relative error in computation of the series terms, modeled by uniform noise
    b=1.
    y=b
    for i in range(1,n+1):
        b*=x/i
        r=2*e*random()-e
        y+=b*(1+r)
    return y
```

```python
x = np.linspace(-3, 2, 50)
y = []
i=0;
for xx in x: 
    y.append(expser(2,xx,.1))
    ++i
plt.plot(x,y)
y = []
i=0;
for xx in x: 
    y.append(expser(100,xx,.1))
    ++i
plt.plot(x,y)
y = []
i=0;
for xx in x: 
    y.append(np.exp(xx))
    ++i
plt.plot(x,y)

plt.show()
```
