```python
import numpy as np
import matplotlib.pyplot as plt
import random as rn
```

```python
def sinbh(x):
#Bhaskara's approximation of sin(x)
    pp=5.*np.pi*np.pi
    z=x*(np.pi-x)
    return 16.*z/(pp-4.*z)
```

```python
def cosbh(x):
#Bhaskara's approximation of cos(x)
    pp=np.pi*np.pi
    xx=x*x;
    return (pp-4.*xx)/(pp+xx)
```

```python
n=20
xlist=np.linspace(-.1, 3.24, 200)
cb=[cosbh(x) for x in xlist]
plt.plot(xlist,cb)
plt.plot(xlist,np.cos(xlist))
plt.show()
```
