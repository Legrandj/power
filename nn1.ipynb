```python
def sigm(x):
    return 1./(1+np.exp(-x))

def dsigm(x):
    s = sigm(x)
    return s*(1-s)

def relu(x):
    return x if x>0 else 0

def drelu(x):
    return 1 if x>0 else 0
```

```python
def pwsigm(f,x,a,h):
    m = 3
    h2 = .5*h
    kk = int(x/h)
    r = 0
    for k in range(kk-m,kk+m):
        kh = k*h
        v = f(kh+h2)
        r += v*sigm(a*(x-kh))
        r -= v*sigm(a*(x-kh-h))
    return r
```

```python
def pxy(a,b,c,X,Y):
    xn, yn = X.shape
    Z = X*0
    for i in range(xn):
        for k in range(yn):
            Z[i,k] = sigm(a*(X[i,k]-b))-sigm(a*(X[i,k]-b-1))+sigm(a*(Y[i,k]-c))-sigm(a*(Y[i,k]-c-1))
    return Z

def spxy(a,b,c,X,Y):
    aa = 20
    bb = 1.5*aa
    xn, yn = X.shape
    Z = X*0
    for i in range(xn):
        for k in range(yn):
            Z[i,k] = sigm(aa*sigm(a*(X[i,k]-b))-aa*sigm(a*(X[i,k]-b-1))+aa*sigm(a*(Y[i,k]-c))-aa*sigm(a*(Y[i,k]-c-1))-bb)
    return Z

def axy(a,b,c,d,X,Y):
    xn, yn = X.shape
    Z = X*0
    for i in range(xn):
        for k in range(yn):
            Z[i,k] = sigm(a*X[i,k]+b*Y[i,k]+c)-sigm(a*X[i,k]+b*Y[i,k]+d)
    return Z
```

```python
x1 = 0
x2 = 6.
y1 = 0.
y2 = 6.
nx = 200
dx = (x2-x1)/nx
ny = nx
dy = (y2-y1)/ny
x = np.arange(x1, x2+dx, dx)
y = np.arange(y1, y2+dy, dy)
X, Y = np.meshgrid(x, y)
Z = pxy(30,3,3,X,Y)
plt.figure(figsize=(10,8))
cvals = np.arange(.001,2.5,.2)
#cvals = np.append(cvals,[2**i for i in np.arange(-1,6,.01)])
plt.imshow(Z, extent=[x1, x2, y1, y2], cmap=plt.cm.gray, origin='lower',aspect='auto')
plt.contour(X, Y, Z,levels=cvals, cmap=plt.cm.rainbow, origin='lower')
plt.colorbar()
plt.xlabel('$x_1$',size=14)
plt.ylabel('$x_2$',size=14)
plt.rcParams['grid.linestyle'] = "--"
plt.rcParams['grid.color'] = 'gray'
#ax.grid(color='r', linestyle='-', linewidth=2)
plt.grid()
#plt.savefig("pxy1.png", dpi=200,bbox_inches='tight')
plt.show()
```
